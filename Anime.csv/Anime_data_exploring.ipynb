```python
import pandas as data
import seaborn as sns
import matplotlib.pyplot as plt


anime_data = data.read_csv("Anime.csv")
anime_data
```

```python
#What is the average number of episodes?

length_TV_types = len(anime_data[anime_data["Type"] == 'TV'])
#length_TV_types

length_of_data = len(anime_data["Type"])
#length_of_types

percent = length_TV_types / length_of_data * 100
percent
```

```python
#What is the average number of episodes?
episodes = anime_data["Episodes"]
total_episodes = episodes.mean()
total_episodes.round()
```

```python
#What is the difference in scores based on the first five anime?
anime_data.head()

data_score = {
    "Name" : ["Attak on Titan", "Death Note", "Fullmetal Alchemist: Brotherhood", "One Punch Man", "Sword Art Online"], "Score" : [8.531, 8.621, 9.131, 8.511, 7.201]
    }

anime_dataframe = data.DataFrame(data_score)

score_bar_plot = sns.barplot(x="Name", y="Score", data=data_score)
score_bar_plot
```
